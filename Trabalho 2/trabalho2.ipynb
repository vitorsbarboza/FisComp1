```python
### Exercício 1 - raiz quadrada de 5 (Método Newton-Raphson)

# Definir a função de x
def f(x):
    y = x**2 - 5
    return y

# Plotar a função
x = np.linspace(-5,5,1000)
y = f(x)
plt.axhline(0)
plt.plot(x,y)
# --> conclusão: a função apresenta dois zeros

# Definir a derivada da função de x
def f1(x):
    y = 2*x
    return y

# Método de Newton-Raphson
E = 1e-10 #precisão
i = 0  #número de iterações
x0 = float(input("Digite o chute inicial: "))
while abs(f(x0)) > E:
    x0 = x0 - f(x0)/f1(x0)
    i += 1
print("O valor de x que satisfaça f(x) = 0 é: ", x0)
print("A quantidade de iterações necessárias foi: ", i)

#Os valores encontrados foram: {-2,23607 ; 2,236067}

```

```python
### Exercício 2

import numpy as np
import matplotlib.pyplot as plt
import time as t

print('------------------------------------------------------------------------------------')
print('Programa para calcular a energia do estado fundamental para um nêutron e um próton')
print('------------------------------------------------------------------------------------')
t.sleep(3)
print("Qual método deseja utilizar? Digite 1 -> Bisseção e 2 -> Secante")
n = int(input())
print('------------------------------------------------------------------------------------')

if n == 1:
  # Método da Bisseção
  V0 = 60
  a = 1.45
  m = 938
  h = 197.326
  a1 = float(input("Digite o mínimo do intervalo: "))
  b1 =float(input("Digite o máximo do intervalo: "))
  E = (a1 + b1)/2
  epsilon = 1E-5
  i = 1

  def f(E):
    b = np.sqrt(m*abs(E))/h
    k = np.sqrt(m*(V0-abs(E)))/h
    y = k/np.tan(k*a) + b
    return y

  x = np.linspace(a1,b1,20000)
  y = f(x)
  plt.plot(x,y)
  plt.xlabel("Energia E (MeV)")
  plt.ylabel("Função transcendental f(E)")
  plt.axhline(0)
  print('------------------------------------------------------------------------------------')

  # Método da Bisseção
  while abs(a1) - abs(b1) > epsilon or abs(f(E)) > epsilon:
    if f(E)*f(b1) < 0:
      a1 = E
    else:
      b1 = E
    E = (a1 + b1)/2
    i += 1

  print("A Energia do Estado Fundamental é E0 = ", E," MeV")
  print("Quantidade de iterações: ", i)
  print(f"Valor da função em para E0: {f(E):.6f}")
  print('------------------------------------------------------------------------------------')


elif n == 2:
  # Método da Secante
  V0 = 60
  a = 1.45
  m = 938
  h = 197.326
  a1 = float(input("Digite o mínimo do intervalo: "))
  b1 =float(input("Digite o máximo do intervalo: "))
  E = a1 - f(a1)*(a1-b1)/(f(a1)-f(b1))
  epsilon = 1E-5
  i = 1

  def f(E):
    b = np.sqrt(m*abs(E))/h
    k = np.sqrt(m*(V0-abs(E)))/h
    y = k/np.tan(k*a) + b
    return y

  x = np.linspace(a1,b1,20000)
  y = f(x)
  plt.plot(x,y)
  plt.xlabel("Energia E (MeV)")
  plt.ylabel("Função transcendental f(E)")
  plt.axhline(0)
  print('------------------------------------------------------------------------------------')

  while abs(abs(a1) - abs(E)) > epsilon or abs(f(E)) > epsilon:
    b1 = a1
    a1 = E
    E = a1 - f(a1)*(a1-b1)/(f(a1)-f(b1))
    i += 1

  print("A Energia do Estado Fundamental é E0 = ", E," MeV")
  print("Quantidade de iterações: ", i)
  print(f"Valor da função para E0: {f(E):.6f}")
  print('------------------------------------------------------------------------------------')
else: 
  print("Digite uma opção válida!")



```
